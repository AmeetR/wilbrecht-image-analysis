```python
subdirs = os.listdir(directory)
tif_list = []
for subdir in subdirs:
    try:
        for file in os.listdir(directory + subdir):
            tif_list += [directory + subdir + "/" + file]
            print(file)
    except Exception as e:
        continue
print(tif_list)
    
```

```python
for i in tif_list:
    index = "{}_{}".format(i[-13:-11],i[-9:-7])
    tif_to_slice[index].append(i)
tif_to_slice
```

```python
path = directory + '/../sectioned_test_1'
try:  
    os.mkdir(path)
except OSError:  
    print ("Creation of the directory %s failed" % path)
else:  
    print ("Successfully created the directory %s " % path)
```

```python
for section in tif_to_slice: 
    try: 
        os.mkdir(path + "/" + section)
    except OSError:  
        print ("Creation of the directory %s failed" % section)
    else:  
        print ("Successfully created the directory %s " % section)
```

```python
for folder in tif_to_slice: 
    source = path + "/" + folder
    for i in tif_to_slice[folder]:
        shutil.copy(i, source)
```
